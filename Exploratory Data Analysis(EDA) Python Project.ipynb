```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"

df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

#converted 0 and 1 values of senior citizen to yes/no to make it easier to understand

```python
ax = sns.countplot(x = 'Churn', data = df)


ax.bar_label(ax.containers[0])
plt.title("Count of Customer by Churn", fontsize = 10)
plt.show()
```

```python
plt.figure(figsize = (4,4))
gb = df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb['Churn'], labels = gb.index, autopct = "%1.2f%%")
plt.title("Percentage of Churned Customers", fontsize = 10)
plt.show()
```

#from the given pie chart we can conclude that 26.54% of our customer have churned out
#now let's explore the reason behind it

```python
plt.figure(figsize = (4,4))
sns.countplot(x = "gender", data =df, hue = "Churn", palette ="Spectral")
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x = "SeniorCitizen", data =df, palette ="seismic")
ax.bar_label(ax.containers[0])
plt.title("Count of Customer by SeniorCitizen")
plt.show()
```

```python
# Count total values for each SeniorCitizen category
total = df["SeniorCitizen"].value_counts().sort_index()

# Count churn values grouped by SeniorCitizen
churn_counts = df.groupby("SeniorCitizen")["Churn"].value_counts(normalize=True).unstack()

# Convert to percentage
churn_counts *= 100

# Plot
fig, ax = plt.subplots(figsize=(4, 4))
churn_counts.plot(kind="bar", stacked=True, colormap="viridis", ax=ax)

# Add labels
for i, bars in enumerate(ax.containers):
    for bar in bars:
        height = bar.get_height()
        if height > 0:  # Only label if bar is non-zero
            ax.text(bar.get_x() + bar.get_width()/2, bar.get_y() + height/2,
                    f"{height:.1f}%", ha='center', va='center', fontsize=10, color="white")

# Labels and title
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage (%)")
plt.title("Churn by Senior Citizen (Stacked % Bar Chart)")
plt.legend(title="Churn")
plt.show()
```

#comparatively a greater percentage of people in senior citizen category have churned

```python
# List of columns to plot
columns = ['PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set the number of rows and columns for subplots
rows, cols = 3, 3  # 3x3 grid

# Create subplots
fig, axes = plt.subplots(rows, cols, figsize=(15, 12))

# Flatten axes array for easy iteration
axes = axes.flatten()

# Loop through each column and plot countplot
for i, col in enumerate(columns):
    sns.countplot(x=df[col], ax=axes[i], hue = df["Churn"], palette="Set2")
    axes[i].set_title(f"Count of {col}")
    axes[i].set_xlabel("")
    axes[i].set_ylabel("Count")

# Adjust layout
plt.tight_layout()
plt.show()
```

#Customers with Fiber Optic Internet and those who lack online security, tech support, or device protection show a higher proportion of churn. On the other hand, services like PhoneService and MultipleLines do not exhibit a strong correlation with churn.
